#Questão 1 (vale até 2 pontos)
Determine a área entre as curvas definidas por $x_1=y^2-2$ e $x_2=y$. Trace o gráfico destas funções sombreando em azul claro a área entre elas. Você pode usar apenas o python, matplotlib, numpy e sympy. Neste caderno do Google Colab.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir as funções
y = sp.Symbol('y')
x1 = y**2 - 2
x2 = y

# Calcular os pontos de interseção entre as curvas
intersection_points = sp.solve(x1 - x2, y)

# Verificar se existem pontos de interseção
if len(intersection_points) == 2:
    # Calcular a área entre as curvas
    area = sp.integrate(sp.Abs(x2 - x1), (y, intersection_points[0], intersection_points[1]))
    area_value = area.evalf()
    print(f"O valor da área entre as curvas é: {area_value}")
else:
    print("Não foram encontrados pontos de interseção.")

# Gerar os valores para o eixo y
y_vals = np.linspace(-10, 10, 400)

# Avaliar as funções nos valores de y
x1_vals = sp.lambdify(y, x1)(y_vals)
x2_vals = sp.lambdify(y, x2)(y_vals)

# Encontrar o intervalo para preencher (entre as curvas)
fill_mask = np.logical_and(x1_vals <= x2_vals, y_vals**2 - 2 <= y_vals)

# Traçar o gráfico e sombrear apenas a área de dentro da curva em azul
plt.figure(figsize=(8, 6))
plt.plot(x1_vals, y_vals, label='x1 = y^2 - 2', color='blue')
plt.plot(x2_vals, y_vals, label='x2 = y', color='red')
plt.fill_betweenx(y_vals, x1_vals, x2_vals, where=fill_mask, color='lightblue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Área entre as curvas x1 = y^2 - 2 e x2 = y')
plt.legend()
plt.grid(True)
plt.show()

```

$F(x) = \int f(x) \, dx = \int (3x^2 - 4x + 2) \, dx$

Aplicando a regra da potência, obtemos:

$F(x) = \frac{3}{3}x^3 - \frac{4}{2}x^2 + 2x + C = x^3 - 2x^2 + 2x + C$

onde $C$ é uma constante de integração.

Calculamos o valor exato da integral definida de $f(x)$ no intervalo $[1, 3]$:

$\int_{1}^{3} f(x) \, dx = F(x) \bigg|_{1}^{3} = F(3) - F(1)$

Substituindo os valores de $x$ na função antiderivada, temos:

$\int_{1}^{3} f(x) \, dx = (3^3 - 2(3)^2 + 2(3)) - (1^3 - 2(1)^2 + 2(1)) = 14$

Portanto, o valor exato da integral definida de $f(x)$ no intervalo $[1, 3]$ é igual a 14.

4. prove que o valor de b que você encontrou produzirá uma área de 6 unidades quadradas com precisão maior ou igual a 0.001.


## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = np.linspace(-10, 10, 100)

y_A = x ** 2
y_B = 2 * x - 1

plt.plot(x, y_A, label='y = x^2')
plt.plot(x, y_B, label='y = 2x - 1')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Curvas A e B')
plt.legend()

plt.show()

x, b = sp.symbols('x b')

y_A = x ** 2
y_B = 2 * x - 1

a = 2

solutions = sp.solve(sp.integrate(y_B - y_A, (x, a, b)) - 6, b)
real_solutions = [sol.evalf() for sol in solutions if abs(sp.im(sol)) < 1e-6]

if len(real_solutions) == 0:
    print("Não foi encontrada nenhuma solução para b.")
else:
    b_value = max(real_solutions)
    print(f"O valor de b será: {b_value}")


    desired_area = 6
    area_expression = sp.integrate(y_B - y_A, (x, a, b_value))
    precision = abs(area_expression - desired_area)

    if precision >= 0.001:
        print("O valor de b não faz com que uma área de 6 unidades quadradas tenha a precisão desejada.")
    else:
        print("O valor de b faz com que uma área de 6 unidades quadradas tenha a precisão desejada.")

x = np.linspace(-10, 10, 100)

y_A = x ** 2
y_B = 2 * x - 1

plt.plot(x, y_A, label='y = x^2')
plt.plot(x, y_B, label='y = 2x - 1')

plt.fill_between(x, y_A, y_B, where=((x >= b_value) & (x <= a)), color='lightblue')

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-2.5, 2.5)
plt.ylim(-5, 5)
plt.title('Curvas A e B com área sombreada')
plt.legend()

plt.show()
```

\end{align*}
